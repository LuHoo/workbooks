## Exercise 1.1 Hypergeometric Distribution

For the hypergeometric model, use `hypergeom.pmf(k, N, M, n)` to calculate point probabilities $P(K = k)$.

To compute cumulative probabilities $P(K \leq k)$, use `hypergeom.cdf`. Right-tail probabilities $P(K > k)$ are obtained with `hypergeom.sf` (the survival function).

```python
N = 331
M = 17
n = 60

hypergeom.pmf(0, N, M, n), hypergeom.pmf(1, N, M, n)
```

```python
hypergeom.cdf(1, N, M, n), hypergeom.pmf(0, N, M, n) + hypergeom.pmf(1, N, M, n)
```

```python
hypergeom.sf(1, N, M, n)
```

## Exercise 1.2 Binomial Distribution

For the binomial approximation, use `binom.pmf` for $P(K = k)$ and `binom.cdf` for $P(K \leq k)$.

The parameters are `k`, `n`, and `p`, where in this case $p = M / N = 17 / 331$.

```python
binom.pmf(k=1, n=60, p=17 / 331), binom.cdf(1, 60, 17 / 331)
```

## Exercise 1.3 Poisson Distribution

Use `poisson.pmf` and `poisson.cdf` with parameter $\mu = n\pi$. For this case, $\mu = 60 \cdot 17/331$.

This gives a compact way to compare individual-event probabilities and cumulative probabilities in the same setting.

```python
mu = 60 * 17 / 331
poisson.pmf(1, mu), poisson.pmf(3, mu)
```

```python
poisson.cdf(1, mu)
```

## Exercise 1.4 Normal Distribution

Use `norm.cdf(x, loc, scale)` with `x` as the observed value, `loc` as the mean, and `scale` as the standard deviation of the sampling distribution.

You can pass arguments positionally or by name. Using named arguments is often clearer for review and makes reordering arguments safe.

```python
scale = 115.26 / math.sqrt(200)
norm.cdf(x=1012, loc=1030, scale=scale)
```

```python
norm.cdf(1012, 1030, scale), norm.cdf(x=1012, loc=1030, scale=scale), norm.cdf(scale=scale, x=1012, loc=1030)
```

## Exercise 1.5 Student's t Distribution

First compute the t-statistic from the sample information, then evaluate the cumulative probability with `t.cdf(x, df)`.

This links the standardized value to its probability under Student's t distribution with the chosen degrees of freedom.

```python
t_val = (1004 - 1030) / (73.8 / math.sqrt(10))
t_val, t.cdf(x=t_val, df=9)
```

## Exercise 1.6 Chi-Squared Distribution

Use `chi2.ppf(q, df)` to obtain critical values (quantiles) of the $\chi^2$ distribution.

Here we compute both an upper quantile (95%) and a lower quantile (5%), which are then used for variance-interval reasoning.

```python
chi2.ppf(q=0.95, df=9), chi2.ppf(q=0.05, df=9)
```

## Exercise 1.7 F Distribution

Use `f.ppf` to get the critical value for hypothesis testing, and `f.sf` to compute the right-tail probability of that statistic.

This mirrors the chapter logic: compare your test statistic to the critical value, then interpret the tail probability in context.

```python
f_crit = f.ppf(q=0.95, dfn=25, dfd=23)
f_crit
```

```python
f.sf(x=f_crit, dfn=25, dfd=23)
```
