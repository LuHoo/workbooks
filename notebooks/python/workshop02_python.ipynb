## Exercise 1.1 Hypergeometric distribution\n
\n
Equivalent to R `dhyper()` / `phyper()`: use `hypergeom.pmf`, `hypergeom.cdf`, and `hypergeom.sf`.

```python
N = 331
M = 17
n = 60

hypergeom.pmf(0, N, M, n), hypergeom.pmf(1, N, M, n)
```

```python
hypergeom.cdf(1, N, M, n), hypergeom.pmf(0, N, M, n) + hypergeom.pmf(1, N, M, n)
```

```python
hypergeom.sf(1, N, M, n)
```

## Exercise 1.2 Binomial distribution\n
\n
Equivalent to R `dbinom()` / `pbinom()`: use `binom.pmf` and `binom.cdf`.

```python
binom.pmf(k=1, n=60, p=17 / 331), binom.cdf(1, 60, 17 / 331)
```

## Exercise 1.3 Poisson distribution\n
\n
Equivalent to R `dpois()` / `ppois()`: use `poisson.pmf` and `poisson.cdf` with `mu = n * pi`.

```python
mu = 60 * 17 / 331
poisson.pmf(1, mu), poisson.pmf(3, mu)
```

```python
poisson.cdf(1, mu)
```

## Exercise 1.4 Normal distribution\n
\n
Equivalent to R `pnorm()`: use `norm.cdf(x, loc, scale)`.

```python
scale = 115.26 / math.sqrt(200)
norm.cdf(x=1012, loc=1030, scale=scale)
```

```python
norm.cdf(1012, 1030, scale), norm.cdf(x=1012, loc=1030, scale=scale), norm.cdf(scale=scale, x=1012, loc=1030)
```

## Exercise 1.5 Student's t distribution\n
\n
Equivalent to R `pt()`: compute `t_val`, then use `t.cdf`.

```python
t_val = (1004 - 1030) / (73.8 / math.sqrt(10))
t_val, t.cdf(x=t_val, df=9)
```

## Exercise 1.7 F distribution\n
\n
Equivalent to R `qf(..., lower.tail = FALSE)` / `pf(..., lower.tail = FALSE)`: use `f.ppf` and `f.sf`.

```python
f_crit = f.ppf(q=0.95, dfn=25, dfd=23)
f_crit
```
